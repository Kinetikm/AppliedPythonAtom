```python
import pandas as pd
```

```python
%%time
data=pd.read_csv('/home/igor/PycharmProjects/vk_bot-hackaton/data/data.csv')
```

```python
gr = data.head(10000).groupby(['sale_date_date','contact_id','shop_id'])
print("yes")
```

```python
import numpy as np
list_of_lists = []
for g in gr.groups:
    #print(len(gr.get_group(g)['product_id'].values))
    if len(gr.get_group(g)['product_id'].values) > 2:
        list_of_lists.append([str(int(i)) for i in gr.get_group(g)['product_id'].values])
#     for item in gr.get_group(g)['product_id'].values:
#         print(item)
#         if type(item) == list and len(item) > 2:
#             list_of_lists.append([str(i) for i in item])
print(list_of_lists)
            
```

```python
from gensim.models import Word2Vec
```

```python
model = Word2Vec(list_of_lists, size=100, window=3, workers=4)
```

```python
model.most_similar(positive=['97117.0', '137539.0'])
```
