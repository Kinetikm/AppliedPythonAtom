```python
# импорт либ
from decision_tree import DecisionTreeClassifier as DTC
from sklearn.tree import DecisionTreeClassifier
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# бинарный случай
X, y = make_classification(n_samples=1000, n_classes=2, n_informative=18, random_state=43)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
MLS=5
```

```python
# учим свою модель
dtc = DTC(min_leaf_size=MLS)
dtc.fit(X_train, y_train.reshape((-1,1)))
y_pred=dtc.predict(X_test)
```

```python
# учим модель из либы
dtc2 = DecisionTreeClassifier(min_samples_leaf=MLS)
dtc2.fit(X_train, y_train.reshape((-1,1)))
y_pred2=dtc2.predict(X_test)
```

```python
# мой скор
accuracy_score(y_test,y_pred)
```

```python
# скор из либы
accuracy_score(y_test,y_pred2)
```

```python
# учим модель из либы
dtc2 = DecisionTreeClassifier(min_samples_leaf=MLS)
dtc2.fit(X_train, y_train.reshape((-1,1)))
y_pred2=dtc2.predict(X_test)
```
